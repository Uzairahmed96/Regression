## Converting Categorical Variables to Numeric by Label Encoding

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
```

```python
ntrain = train.shape[0]
y = train['loss']
test_id = test['id']
data = pd.concat([train.drop(['loss', 'id'], axis = 1), test.drop(['id'], axis=1)], axis = 0)
print(data.shape)
data.head()
```

```python
cols = data.dtypes[data.dtypes == 'object']


for c in cols.index:
    lbl = LabelEncoder() 
    data[c] = lbl.fit_transform(list(data[c].values))
```

```python
data.head()
```

## Train Test Split And Feature Selection
In this dataset, we have a lot of columns with extremely low variance. I will simply remove these variables. We have 2 highly correlated numerical features, in order to fix multicollinearity, I will remove one of the feature.

```python
X = data.drop(['cat3', 'cat7','cat8', 'cat11', 'cat13', 'cat14', 'cat15', 'cat16',
              'cat17', 'cat18', 'cat19', 'cat20', 'cat21', 'cat22', 'cat24',
              'cat26', 'cat28', 'cat29', 'cat30', 'cat31', 'cat32', 'cat33', 'cat34',
              'cat35', 'cat39', 'cat40', 'cat41', 'cat42', 'cat43', 'cat45', 'cat46',
              'cat47', 'cat48', 'cat49', 'cat51', 'cat52', 'cat54', 'cat55', 'cat56',
              'cat57', 'cat58', 'cat59', 'cat60', 'cat61', 'cat62', 'cat63', 'cat64',
              'cat65', 'cat66', 'cat67', 'cat68', 'cat69', 'cat70', 'cat71', 'cat74',
              'cat76', 'cat77', 'cat78', 'cat85', 'cat89', 'cat90', 'cat102', 'cont11'], axis=1)

train_val = X[:ntrain]
test = X[ntrain:]
log_y = np.log(y + 1)
```

# Machine Learning

```python
import xgboost as xgb
```

## Xgboost

```python
trainset = xgb.DMatrix(data = train_val, label = log_y)
```

```python
params = {}
params['objective'] = 'reg:squarederror'
params['learning_rate'] = 0.02
params['booster'] = 'gbtree'
params['subsample'] = 0.90
params['colsample_bytree'] = 0.30
params['max_depth'] = 7
params['gamma'] = 0.5
params['min_child_weight'] = 1
params['reg_alpha'] = 0.31
params['reg_lambda'] = 0.36

model_xgb = xgb.cv(params = params, 
                   dtrain = trainset,
                   num_boost_round=4000,
                   nfold = 5,
                   maximize = False,
                   shuffle = True,
                   seed = 42,
                   verbose_eval = True, 
                   early_stopping_rounds=10,
                   metrics = 'mae')
```

```python
model = xgb.train(params = params, 
                  dtrain=trainset,
                  num_boost_round=model_xgb.index[-1])
```
